```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image=cv2.imread('/content/page1.jpg')
plt.imshow(image)
```

```python
y_coordinate=[
    135,
    191,
    295,
    358,
    421,
    533,
    633,
    691,
    753,
    826,
    963,
    1033
] #from OCR free API key.
```

```python
n = len(y_coordinate)-1
i=0
cropped_images=[]
while i<n:
  cropped_image=image[y_coordinate[i]-5: y_coordinate[i+1]+5,0:1500]
  i=i+1
  cropped_images.append(cropped_image)
```

```python
for i in cropped_images:
  plt.imshow(i)
  plt.show()
```

```python
import os

# Create a directory to store the cropped images.
output_dir = 'cropped_images'
os.makedirs(output_dir, exist_ok=True)

# Save each cropped image to the directory.
for i, cropped_image in enumerate(cropped_images):
  cv2.imwrite(os.path.join(output_dir, f'cropped_image_{i}.jpg'), cropped_image)
```
